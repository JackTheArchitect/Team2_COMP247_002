```python
def make_pie(features, figsize=(14,10)):

    features_length = len(features)

    fig, axs = plt.subplots(features_length, 2, figsize=figsize)
    
    for i in range(features_length):
        
        current_feature = features[i]
        
        stolen_cats = bike_theft_df_stolen[current_feature].value_counts().index
        recovered_cats = bike_theft_df_recovered[current_feature].value_counts().index
        
        axs[i, 0].pie(bike_theft_df_stolen[current_feature].value_counts(), labels=stolen_cats, autopct='%1.1f%%')
        axs[i, 0].set_title(f"{current_feature} - Stolen")
        
        axs[i, 1].pie(bike_theft_df_recovered[current_feature].value_counts(), labels=recovered_cats, autopct='%1.1f%%')
        axs[i, 1].set_title(f"{current_feature} - Recovered")
        
        
    plt.show()
```

```python
def make_bars_2(feature, df=bike_theft_df, data_range=None, figsize=(14,10), title=None):
    stolen = bike_theft_df[bike_theft_df['Status']=='STOLEN'][feature].value_counts()
    recovered = bike_theft_df[bike_theft_df['Status']=='RECOVERED'][feature].value_counts()
    
    if data_range is not None:
        stolen = stolen[:data_range]
        recovered = recovered[:data_range]
        
    if title is None:
        title = feature

    plt.figure(figsize=figsize)
    figure1 = plt.bar(range(len(stolen)), stolen, align='edge', width=0.4, label='Stolen')
    figure2 = plt.bar(range(len(recovered)), recovered, align='edge', width=-0.4, label='Recovered')
    plt.legend()
    plt.xticks(np.arange(len(stolen)), stolen.index)
    plt.title(title)
```

# 2. Data Modeling

```python
from sklearn.utils import resample

def balance_upsample_downsample(n_balance=None, random_state=123):
    df_majority = bike_theft_df[bike_theft_df.Status == 'STOLEN'] 
    df_minority = bike_theft_df[bike_theft_df.Status == 'RECOVERED'] 
    
    if n_balance is None:
        n_balance = len(df_majority)
    
    df_minority_upsampled = resample(df_minority, 
                                     replace=True,     
                                     n_samples=n_balance,    
                                     random_state=random_state)

#     df_majority_downsampled = resample(df_majority, 
#                                      replace=False,    
#                                      n_samples=n_balance,    
#                                      random_state=random_state)
    
    return pd.concat([df_minority_upsampled, 
                     df_majority])
```

```python
balanced_df = balance_upsample_downsample()
balanced_df['Status'].value_counts()
```

```python
# Feature Selection

features_df = balanced_df[["Primary_Offence", "diff", "Premises_Type", "Cost_of_Bike", "Occurrence_DayOfWeek", "Occurrence_Year", "Report_Hour", "Occurrence_Month", "Location_Type", "NeighbourhoodName", "Longitude", "Latitude"]]
labels_df = balanced_df[["Status"]]
pp_labels_df = (labels_df.Status == "RECOVERED").astype(int)
```

```python
pp_labels_df.value_counts()
```

```python
# Train, Test Data spliting
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features_df, pp_labels_df, test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
```

```python
stdSlr = StandardScaler()
ohEdr = OneHotEncoder(handle_unknown='ignore', sparse=False)
simImpNum = SimpleImputer(missing_values=np.nan, strategy='mean')
simImpCat = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
```

```python
num_pipe = Pipeline([
    ('imputer', simImpNum),
    ('scaler', stdSlr),
])

cat_pipe = Pipeline([
    ('imputer', simImpCat),
    ('encoder', ohEdr)
])
```

```python
numeric_df = list(features_df.select_dtypes(include=[np.number]).columns)
cat_df = list(features_df.select_dtypes(exclude=[np.number]))
```

```python
preprocessor = ColumnTransformer([("numeric", num_pipe, numeric_df),
                                  ("categoryical", cat_pipe, cat_df)])
```

```python
transformer_pipeline = Pipeline([
    ('preprocessor', preprocessor)
])

X_prepared = transformer_pipeline.fit_transform(X_train)
```

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state=42, loss="modified_huber", max_iter=5000, tol=1e-4)

sgd_clf.fit(X_prepared, y_train)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(sgd_clf, X_prepared, y_train, cv=3, scoring="accuracy").mean()
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import precision_score, recall_score, confusion_matrix

y_train_pred = cross_val_predict(sgd_clf, X_prepared, y_train, cv=3)

print(precision_score(y_train, y_train_pred))
print(recall_score(y_train, y_train_pred))
print(confusion_matrix(y_train, y_train_pred))
```
